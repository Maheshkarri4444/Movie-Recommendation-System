```python
import numpy as np
import pandas as pd
```

```python
movies_data=pd.read_csv("movies.csv")
movies_data.head(2)
```

```python
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies_data['homepage'].isnull().sum()
```

```python
movies_data.shape
```

```python
selected_features=['genres','keywords','tagline','cast','director','overview']
selected_features
```

```python
for feature in selected_features:
    print(movies_data[feature].isna().sum())
```

```python
for feature in selected_features:
    movies_data[feature]=movies_data[feature].fillna('')
```

```python
combined_features=movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']+' '+movies_data['overview']
```

```python
combined_features
```

```python
#converting the text data to feature vectors
vectorizer=TfidfVectorizer()
```

```python
feature_vectors= vectorizer.fit_transform(raw_documents=combined_features)
```

```python
print(feature_vectors)
```

```python
similarity=cosine_similarity(feature_vectors)
similarity
```

```python
similarity.shape
```

```python
titles_list=movies_data['title'].tolist()
movies_list=movies_data[['index','title','id']].values.tolist()
```

```python
titles_list
```

```python
movie_name="iron man"
find_the_movie=difflib.get_close_matches(movie_name,titles_list,cutoff=0.5)
```

```python
print(find_the_movie)
```

```python
close_match=find_the_movie[0]
close_match 
```

```python
# find the index of the movie with title
index_of_the_movie=movies_data[movies_data.title==close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
similarity_score=list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#sorting the similarity scores
sorted_similar_movies=sorted(similarity_score,key= lambda x:x[1],reverse=True)
```

```python
sorted_similar_movies
```

```python
#printing the names of similer movies
print('Movies Suggested for you : \n')
i=1
for movie in sorted_similar_movies:
    index=movie[0]
    title_from_index=movies_data[movies_data.index==index]['title'].values[0]
    if i<11:
        print(i,'.',title_from_index)
        i+=1
```

```python
def recommend(name):
    finding=difflib.get_close_matches(name,titles_list,cutoff=0.5)
    if(len(finding)==0):
       print("The Movie is not found in our Dataset \n")
       print("please enter a hollywood movie!")
    else:
        matching=finding[0]
        index_movie=movies_data[movies_data.title==matching]['index'].values[0]
        sim_score=list(enumerate(similarity[index_movie]))
        sorted_sim_movies=sorted(sim_score,key= lambda x:x[1],reverse=True)
        print("Movies recommended for you : ")
        i=1
        for movie in sorted_sim_movies:
            index=movie[0]
            title_from_index=movies_data[movies_data.index==index]['title'].values[0]
            if i<11:
                print(i,'.',title_from_index)
                i+=1       
```

```python
recommend('avatar')
```

```python
recommend('mission impossible')
```
